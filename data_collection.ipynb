```python
import lyricsgenius as genius
import pickle
```

I am using LyricsGenius API described $\href{https://github.com/johnwmillr/LyricsGenius}{here}$ to get top 100 song lyrics (by popularity) and their translations from $\href{https://genius.com/}{genius.com}$

Then I collect all translations and save it as a pickle file. This file will be used in the other notebook.

```python
artist = 'BTS'
max_songs = 100
access_token = '1_cCky2Ywyz79M4_3zuNQrDMQHF8TdLnZ7A7rMivBnMYAyJWvNIxcs0EcazGFirq'
artist = genius.Genius(access_token).search_artist(artist, max_songs=max_songs, sort = 'popularity')
titles = ['BTS - ' + song.title + ' (English Translation)' for song in artist.songs]
```

```python
songs = []
for title in titles:
    song = genius.Genius(access_token).search_song(title, 'Genius Translations')
    if song not in songs:
        songs.append(song)
songs = [s for s in songs if s is not None]
song_pickle = open("test.pickle", "wb")
pickle.dump(songs, song_pickle)
song_pickle.close()
```
